```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
df['FlightNumber']
```

```python
new_index=np.arange(1,df.From_To.count()+1)
new_index
df.set_index(new_index, inplace=True)
df
```

```python
for i in np.arange(1,df.From_To.count()+1):
    if pd.isnull(df.FlightNumber.loc[i,]):
        df.loc[i,'FlightNumber'] = df.FlightNumber.loc[i-1,] + 10
df['FlightNumber']
df
```

```python
df['FlightNumber'].astype(int)
```

```python
df1 = df.copy()
df1[['From','To']] = df1.From_To.str.split("_",expand=True)
df1
```

```python
df1.From = df1.From.str.capitalize()
df1.To = df1.To.str.capitalize()
df1.From_To = df1.From_To.str.capitalize()
print(df1)
```

```python
df.drop('From_To',axis=1,inplace=True)
df
```

```python
df['From_To'] = df1['From_To']
df
```
